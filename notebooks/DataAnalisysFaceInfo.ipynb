### Import Libs ###

```python
from sklearn.utils import class_weight
import pandas as pd
import numpy as np
```

### Load data ###

```python
df = pd.read_csv("fairface_label_new_train.csv")
df.head()
```

```python
df.race.value_counts().plot.bar();
```

```python
race_weights = class_weight.compute_class_weight('balanced', np.unique(df.race), df.race)
{x:y for x, y in zip(np.unique(df.race), race_weights)}
```

```python
df.gender.value_counts().plot.bar();
```

```python
gender_weights = class_weight.compute_class_weight('balanced', np.unique(df.gender), df.gender)
{x:y for x, y in zip(np.unique(df.gender), gender_weights)}
```

```python
df.age.value_counts().plot.bar();
```

```python
age_weights = class_weight.compute_class_weight('balanced', np.unique(df.age), df.age)
{x:y for x, y in zip(np.unique(df.age), age_weights)}
```
